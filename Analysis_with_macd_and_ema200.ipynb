```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('sh000001.csv',index_col='date')
```

```python
data
```

```python
data['DIF']=data['close'].ewm(span=12).mean()-data['close'].ewm(span=26).mean()
```

```python
data['DEA']=data['DIF'].ewm(span=9).mean()
```

```python
data['EMA200']=data['close'].ewm(span=200).mean()
```

```python
# 该列辅助判断当前是在EMA200以上还是以下
data['EMA200-close']=data['EMA200']-data['close']
```

```python
data
```

## 结合EMA200复盘金叉买、死叉卖的收益结果

```python
data['DIF-DEA']=data['DIF']-data['DEA']
```

```python
data
```

```python
# data['div_temp']为判断金叉死叉辅助列
data['div_temp']=data['DIF-DEA']/data['DIF-DEA'].shift(1)
```

```python
data
```

```python
data=data.dropna()
```

```python
data
```

```python
data['res']=data.apply(lambda x: 0 if (x['div_temp'] >= 0) else (1 if (x['DIF-DEA'] > 0) else -1), axis=1)
```

```python
data
```

```python
data=data[~data['res'].isin([0])]
data
```

```python
data['EMA200-close']=data.apply(lambda x:(1) if(x['EMA200-close']>0) else -1,axis=1)
```

```python
data
```

```python
data=data[data['EMA200-close']==data['res']]
data
```

```python
data['res']=data['res']*data['res'].shift(1)
```

```python
data
```

```python
data=data[~data['res'].isin([1])]
```

```python
data
```

```python
data['mutply']=data['open']
```

```python
data['mutply']=data.apply(lambda x:(1/x['mutply']) if(x['EMA200-close']==1) else x['mutply'],axis=1)
data
```

```python
return_final=data['mutply'].cumprod()
```
